### Test random image of an bird from cifar10 Dataset 

```
# Load VGG16 pretrained model on imagenet data
model = VGG16(weights="imagenet")
# Pick a random unnormalized and unstandarized test image. i.e in this case say 25th test image
testimage1 = test_features[25]
#Since VGG16 has fully connected layers the image sizes used for predicions should have the same size of training images. (i.e 224*224*3)
testimage1 = cv2.resize(testimage1, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
x = image.img_to_array(testimage1)

#Using expand_dims to convert a image into a batch of size 1
x = np.expand_dims(x, axis=0)
#Preprocessing the raw test image to have same preprocessing as required for the trained model which normalizes and standardizes the given raw test image
x = preprocess_input(x)
```

1. We get the predictions for the image
2. We take the topmost class index
3. Take output from the final convolutional layer

```
model.summary()
```

### Gradcam at a layer one before block5_conv3 that is on layer block5_conv2

```
preds = model.predict(x)
class_idx = np.argmax(preds[0])
print(class_idx)
class_output = model.output[:, class_idx]
last_conv_layer = model.get_layer("block5_conv2")
```

1. Compute the gradient of the class output value with respect to the feature map
2. Pool the gradients over all the axies leaving out the channel dimension
3. Weigh the output feature map with the computed gradient values

```
grads = K.gradients(class_output, last_conv_layer.output)[0]
print(grads.shape)
pooled_grads = K.mean(grads, axis=(0, 1, 2))
print(pooled_grads.shape)
iterate = K.function([model.input], [pooled_grads, last_conv_layer.output[0]])

pooled_grads_value, conv_layer_output_value = iterate([x])

for i in range(512):
  conv_layer_output_value[:, :, i] *= pooled_grads_value[i]
```

1. Average the weighted feature map along the channel dimension resulting in a heat map of size 14x14 
2. Normalize the heat map to make the values between 0 and 1

```
heatmap = np.mean(conv_layer_output_value, axis = -1)
print(conv_layer_output_value.shape)
print(heatmap.shape)
heatmap = np.maximum(heatmap, 0)
heatmap /= np.max(heatmap)
```

Show the results

```
heatmap = cv2.resize(heatmap, (testimage1.shape[1], testimage1.shape[0]))
heatmap = np.uint8(255 * heatmap)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
superimposed_img = cv2.addWeighted(testimage1, 0.5, heatmap, 0.5, 0)
from google.colab.patches import cv2_imshow
cv2_imshow( testimage1)
cv2_imshow(superimposed_img)
```




1.	The network Architecure  has 1 convolution followed by 3 contracting layers followed by 3 Expanding layers.
2.	The first convolution is a regular 3x3x3 Conv with stride as 1, using receptive field formula, we get RF as 3x3x3
3.	After that  we have a conv layer with stride as 2, Rf calculation now is 3 + 1*(3-1) = 5
4.	After that  we have a normal conv with stride 1, RF calculation now is 5 + 2*(3-1) = 9
5.	Now we have a conv layer with stride as 2, Rf calculation now is 9 + 2*(3-1) = 13
6.Then we have a normal conv with stride 1, RF calculation now is 13 + 4*(3-1) = 21
7.	 we have a conv layer with stride as 2, Rf calculation now is 21 + 4*(3-1) = 29
8.	Now we have a normal conv with stride 1, RF calculation now is 29 + 8 * (3-1) = 45
